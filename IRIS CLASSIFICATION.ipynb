```python
from sklearn import datasets, svm, metrics
from sklearn.model_selection import train_test_split

# Load the iris dataset
iris = datasets.load_iris()

# Split the dataset into features (X) and labels (y)
X = iris.data
y = iris.target

# Split the data into training and testing sets
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2)

# Separate the features into Petal and Sepal features
train_x_p = train_X[:, 2:]
test_x_p = test_X[:, 2:]
train_x_s = train_X[:, :2]
test_x_s = test_X[:, :2]

# Train the SVM models

# Model using Petals
model_petals = svm.SVC()
model_petals.fit(train_x_p, train_y)
prediction_petals = model_petals.predict(test_x_p)
accuracy_petals = metrics.accuracy_score(prediction_petals, test_y)
print('The accuracy of the SVM using Petals is:', accuracy_petals)

# Model using Sepals
model_sepal = svm.SVC()
model_sepal.fit(train_x_s, train_y)
prediction_sepal = model_sepal.predict(test_x_s)
accuracy_sepal = metrics.accuracy_score(prediction_sepal, test_y)
print('The accuracy of the SVM using Sepals is:', accuracy_sepal)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Define the models
models = {
    'KNN': KNeighborsClassifier(),
    'SVC': svm.SVC(),
    'Decision Tree': DecisionTreeClassifier()
}

# Train and evaluate each model
accuracies = []
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    # Store the accuracy
    accuracies.append((name, accuracy))

# Plot the graph
labels, scores = zip(*accuracies)
fig, ax = plt.subplots()
ax.bar(labels, scores)
ax.set_xlabel('Algorithm')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy Scores for Various Algorithms')
plt.show()
```
